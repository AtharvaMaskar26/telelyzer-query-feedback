```python
DATA1 = "data\Telelyzer V2.02 Markings - telelyzer-v2.3-markings.csv"
DATA2 = "data\Telelyzer V2.02 Markings - telelyzer-v3.1-markings.csv"
```

```python
import pandas as pd
```

```python
data1 = pd.read_csv(DATA1)
data2 = pd.read_csv(DATA2)
```

```python
data2
```

```python
import json
```

```python
import ast

qa = []

for index, row in data1.iterrows():
    question_answer_pairs = ast.literal_eval(row['Query Solution Pairs'])
    for qa_pair in question_answer_pairs:
        qa.append(
            qa_pair
        )

for index, row in data2.iterrows():
    question_answer_pairs = ast.literal_eval(row['Query Solution Pairs'])['query_solution_pairs']
    for qa_pair in question_answer_pairs:
        print(qa_pair)
        qa.append(
            qa_pair
        )
```

```python
with open("./data/question_answer_pairs.json", "w") as jsonFile:
    json.dump(qa, jsonFile, indent=4)
```

```python
qa_pairs = []
```

```python
for i in qa:
    if "actual_answer" in i.keys():
        qa_pairs.append(
            {"query": i['query'], 
            "solution": i['agent_solution']}
        )

        qa_pairs.append(
            {"query": i['query'],
            "solution": f"(System Solution): {i['agent_solution']}"}
        )
    else:
        qa_pairs.append(
            {"query": i['query'], 
            "solution": i['agent_solution']}
        )
```

```python
with open("./data/question_solution_pairs.json", "w") as jsonFile:
    json.dump(qa_pairs, jsonFile, indent=4)
```
